## SINDy Formulation

We begin by describing the **Sparse Identification of Nonlinear Dynamical Systems (SINDy)** framework.

Consider a dynamical system of the form:

$$
\dot{x} = f(x), \quad t > 0
$$

The goal of SINDy is to identify a parsimonious model that approximates the underlying dynamics by performing sparse regression on a set of **candidate library functions** collected in the matrix $\Theta(x)$.  
The model can be written as:

$$
\dot{x} \approx \Theta(x)\mathbf{W},
$$

where:
- $\Theta(x)$ is a library of candidate nonlinear functions (e.g., polynomials, trigonometric terms),
- $\mathbf{W}$ is a matrix of coefficients representing the contribution of each term to the dynamics.

Suppose we have a set of sampled data points $\{(x_i, \dot{x}_i)\}_{i=1}^{N}$, which can be compactly written as the matrices $X$ and $\dot{X}$.  
Then, the identification problem becomes a sparse regression problem:

$$
\mathbf{W} = \arg\min_{W} \Big( \|\dot{X} - \Theta(X)W\|_2^2 + \lambda \|W\|_0 \Big),
$$

where:
- the first term enforces **data fidelity** (least-squares fit),
- the second term promotes **sparsity** in the coefficient matrix $\mathbf{W}$,
- $\lambda$ is a regularization parameter controlling the trade-off between accuracy and sparsity.

This formulation allows one to identify the most relevant terms in the dynamics, leading to interpretable and compact models.

## Noisy Data and the Weak SINDy Formulation

We now consider the case in which the observed data are corrupted by noise.  
Let
$$
x_i = x_i^{\text{true}} + \eta_i, \qquad \eta_i \sim \mathcal{N}(0, \sigma_i),
$$
where we allow for **heteroscedastic noise**, that is, $\sigma_i \neq \sigma_j$ in general.

The presence of noise has a particularly strong impact on the estimation of derivatives, often amplifying errors and leading to instability in the recovered dynamics.  

#### Weak Formulation

To mitigate this issue, we adopt the **weak formulation**, which avoids computing explicit derivatives from noisy trajectories.

We introduce a family of **test functions** $\phi(t)$ and multiply both sides of the dynamical system by $\phi(t)$ before integrating over time:
$$
\int_0^T \dot{x}(t)\, \phi(t)\, dt = \int_0^T f(x(t))\, \phi(t)\, dt.
$$

Applying **integration by parts** to the left-hand side yields:
$$
- \int_0^T x(t)\, \dot{\phi}(t)\, dt = \int_0^T f(x(t))\, \phi(t)\, dt,
$$
which defines the **weak form** of the dynamical system.

This formulation transfers differentiation from the (noisy) data $x(t)$ to the smooth, user-defined test functions $\phi(t)$, significantly improving robustness in the presence of measurement noise.

#### Matrix Formulation

We now define the **Gram–Schmidt (test function) matrices** $ V $ and $ \dot{V} $, whose entries are temporal inner products between the data and the test functions (or their derivatives):

$$
V_{ij} = \int_0^T \phi_j(t)\, \phi_i(t)\, dt,
\qquad
\dot{V}_{ij} = -\int_0^T \dot{\phi}_j(t)\, \phi_i(t)\, dt.
$$

Using these matrices, we can express the weak formulation of the dynamics compactly as

$$
\dot{V}\, X \approx V\, \Theta(X)\, w,
$$

Rearranging yields the standard **linear regression form**

$$
b = G\, w,
$$

where
$$
b = \dot{V}\, X, \qquad G = V\, \Theta(X).
$$

Here:
- $ b $ is the **measurement vector** obtained from weak differentiation,
- $ G $ is the **design (Gram) matrix** containing weak features,
- and $ w $ represents the unknown coefficients of the governing equations.
